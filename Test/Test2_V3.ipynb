```python
import pandas as pd
import numpy as np
from efficient_apriori import apriori

dataset=pd.read_csv('./订单表.csv',encoding="gbk")
dataset=dataset.sort_values(by='订单日期')
dataset.reset_index(inplace=True)
del dataset['index']
datastep1=dataset.loc[:,['订单日期','产品名称']]
#print(datastep1.head())
datastep2=datastep1.drop_duplicates()
#print(datastep2.head())
#datastep3=datastep2.pivot(columns='产品名称',values='产品名称')
datastep4=datastep2.pivot(index='订单日期',columns='产品名称',values='产品名称')
#print(datastep3.head())
print(datastep4.head())
#print(datastep1.shape[0])
```

```python
transactions = []
for i in range(0,datastep4.shape[0]):
    temp=[]
    for j in range(0,datastep4.shape[1]):
        # print(dataset.values[i,j])
        if str(datastep4.values[i,j]) !='nan':
            temp.append(str(datastep4.values[i,j]))
    transactions.append(temp)
```

```python
itemsets, rules = apriori(transactions,min_support=0.5,min_confidence=0.8)
print('使用efficient_apriori工具包','-'*20)
print('频繁项集：',itemsets)
print('关联规则：',rules)
```

```python
itemsets, rules = apriori(transactions,min_support=0.3,min_confidence=0.5)
print('使用efficient_apriori工具包','-'*20)
print('频繁项集：',itemsets)
print('关联规则：',rules)
```
